```python
# Mengubah data menjadi data frame
df = pd.DataFrame(data)
df
```

```python
#Menentukan label
train_data = df[df['diabetes'].notnull()]
test_data = df[df['diabetes'].isnull()]

X_train = train_data[['gula_darah', 'tekanan_darah']]
y_train = train_data['diabetes']

X_test = test_data[['gula_darah', 'tekanan_darah']]
```

```python
# Mengimplementasikan KNN
# nilai k = 3
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
```

```python
# Memprediksi nilai daiabetes untuk test data
y_pred = knn.predict(X_test)
```

```python
# Update dataframe dengan hasil prediksi
df.loc[df['diabetes'].isnull(), 'diabetes'] = y_pred

# Menapilkan dataframe
print(df)
```
